## Decision Tree

```python
#Importing library for decision tree
from sklearn.tree import DecisionTreeRegressor
```

```python
dt = DecisionTreeRegressor(random_state=42, max_depth=4, min_samples_leaf=10)
```

```python
from sklearn.model_selection import train_test_split
np.random.seed(0)
df_train, df_test = train_test_split(df, train_size=0.7, random_state=100)
```

```python
df_train.shape, df_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
y_train = df_train.pop("Species")
X_train = df_train

y_test = df_test.pop("Species")
X_test = df_test
```

```python
# Defining the decision tree algorithm
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier(random_state=0)
dtree.fit(X_train,y_train)

print('Decision Tree Classifer Created')
```

```python
# predicting on test dataset
y_pred=dtree.predict(X_test)
y_pred
```

```python
Feature_names=list(df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']])
```

```python
from sklearn import tree
Tree=plt.figure(figsize=(22,10))
Tree=tree.plot_tree(dtree,feature_names=Feature_names,fontsize=14,filled='True',proportion=True,rounded=10)
plt.show()
```

```python
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score

print("Model Accuracy is : ",round(accuracy_score(y_pred,y_test)*100,2))

print(classification_report(y_pred,y_test))
```
